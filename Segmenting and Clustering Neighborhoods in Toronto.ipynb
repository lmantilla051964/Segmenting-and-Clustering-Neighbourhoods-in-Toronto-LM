```python
##Scraping the Wikipedia page for the table of Canada's postal codes
```

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
##The html table is converted to Pandas DataFrame for cleaning and preprocessing
```

```python
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

```python
# Dropping the rows where Borough is 'Not assigned' - as requested by 3.2 Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.
Toronto = df[df.Borough != 'Not assigned']

# Combining the neighbourhoods with same Postalcode - 3.3 More than one neighborhood can exist in one postal code area - combine the rows

Toronto.groupby('Postal Code').agg({'Borough':'first','Neighbourhood': ', '.join}).reset_index()
                      
                         
# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
Toronto['Neighbourhood'] = np.where(Toronto['Neighbourhood'] == 'Not assigned',Toronto['Borough'], Toronto['Neighbourhood'])

Toronto
                         
```

```python
# Shape of data frame
Toronto.shape
```

```python
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head(30)
```

```python
##Merging the two tables for getting the Latitudes and Longitudes for various neighbourhoods
```

```python
lat_lon.rename(columns={'Postal Code':'Postal Code'},inplace=True)

Toronto = pd.merge(Toronto,lat_lon,on='Postal Code')
Toronto.head()

#lat_lon.rename(columns={'Postal Code':'Postcode'},inplace=True)
#df3 = pd.merge(df2,lat_lon,on='Postcode')
#df3.head()
```
